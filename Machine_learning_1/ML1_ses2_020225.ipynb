```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/skathirmani/datasets/refs/heads/main/insurance.csv"
```

```python
insurance = pd.read_csv(url)
```

```python
insurance
```

```python
cat_cols=['sex','smoker','region']
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder
one_encoder=OneHotEncoder(sparse_output=False).fit(insurance[cat_cols])
encoded_data=one_encoder.transform(insurance[cat_cols])
encoded_data
```

```python
## Converting Encoder into data frame
df_encoded = pd.DataFrame(encoded_data,columns=one_encoder.get_feature_names_out())
df_encoded
```
